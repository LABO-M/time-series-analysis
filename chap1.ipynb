### 正規分布

- stats.norm($\mu$,$\sigma$)で正規分布のインスタンス(例えばrv)を生成

- rv.rvs(n),rv.rvs(n,m): サイズ$n,n\times m$の正規乱数


- rv.mean(),rv.var,rv.pdf(),rv.cdf(),rv.isf():確率変数の期待値、分散、確率密度、累積分布、上側％点を計算。

```python
rv = stats.norm(0,1)
rv.mean(),rv.var(),rv.pdf(0),1/np.sqrt(2*np.pi),rv.isf(0.01)
```

正規ホワイトノイズ

```python
rv_mu_0_sigma_1= stats.norm(0,1)# 平均0,標準偏差1の正規分布
xt_mu_0_sigma_1= rv_mu_0_sigma_1.rvs(100)# 100個の乱数を生成
rv_mu_2_sigma_1= stats.norm(2,1)
xt_mu_2_sigma_1= rv_mu_2_sigma_1.rvs(100)
rv_mu_n2_sigma_1=stats.norm(-2,1)
xt_mu_n2_sigma_1= rv_mu_n2_sigma_1.rvs(100)
rv_mu_0_sigma_2=stats.norm(0,2)
xt_mu_0_sigma_2= rv_mu_0_sigma_2.rvs(100)
rv_mu_0_sigma_3=stats.norm(0,3)
xt_mu_0_sigma_3= rv_mu_0_sigma_3.rvs(100)
rv_mu_2_sigma_2=stats.norm(2,2)
xt_mu_2_sigma_2= rv_mu_2_sigma_2.rvs(100)
ts = np.arange(1,101,1)
```

```python
plt.plot(ts,xt_mu_0_sigma_1)
plt.title("(a) $\mu=0,\sigma=1$")
plt.xlabel("t")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(ts,xt_mu_2_sigma_1)
plt.title("(b) $\mu=2,\sigma=1$")
plt.xlabel("t")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(ts,xt_mu_n2_sigma_1)
plt.title("(c) $\mu=-2,\sigma=1$")
plt.xlabel("t")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(ts,xt_mu_0_sigma_2)
plt.title("(d) $\mu=0,\sigma=2$")
plt.xlabel("t")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(ts,xt_mu_0_sigma_3)
plt.title("(e) $\mu=0,\sigma=3$")
plt.xlabel("t")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(ts,xt_mu_2_sigma_2)
plt.title("(f) $\mu=2,\sigma=2$")
plt.xlabel("t")
plt.ylabel("y")
plt.show()
```
